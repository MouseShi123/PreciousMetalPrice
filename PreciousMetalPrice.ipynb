```
import pandas as pd
import pandas_datareader as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
!pip install yfinance
import yfinance as yf
start = datetime.datetime(2001,1,1)
start1 = datetime.datetime(2016,1,1)
start2 = datetime.datetime(2020,1,1)
end = datetime.date.today()
```

```
data_meta0 = yf.download("GC=F SI=F PL=F PA=F", start=start, end=end)
data_meta2 = yf.download("GC=F SI=F PL=F PA=F", start=start2, end=end)

fig, axs = plt.subplots(4,figsize=(16,10))
axs[0].plot(data_meta0['Close']["GC=F"])
axs[1].plot(data_meta0['Close']["SI=F"])
axs[2].plot(data_meta0['Close']["PL=F"])
axs[3].plot(data_meta0['Close']["PA=F"])
axs[0].set_title('Gold')
axs[1].set_title('Silver')
axs[2].set_title('Platinum')
axs[3].set_title('Palladium')
fig.tight_layout()
```
